Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
#A scatter plot, also known as a scattergram, scatter graph, or scatter diagram, is a type of graph used to display the relationship between two numerical variables. Here’s how it works:

#Each data point in the dataset is plotted as a point on the graph.
#The x-coordinate of each point represents the value of one variable.
#The y-coordinate of each point represents the value of another variable.
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

answer:- 
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots (axes) in rows and columns, making it easier to compare and visualize different data simultaneously123. Each subplot can represent a different aspect of your data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)  # Create a 2x2 grid, select position 1
plt.plot(x, y1, label="Line 1")
plt.title("Line 1")

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)  # Select position 2
plt.plot(x, y2, label="Line 2")
plt.title("Line 2")

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)  # Select position 3
plt.plot(x, y3, label="Line 3")
plt.title("Line 3")

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)  # Select position 4
plt.plot(x, y4, label="Line 4")
plt.title("Line 4")

# Add common labels and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Adjust layout for better readability
plt.tight_layout()

# Show the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

answer:-4 A bar plot, also known as a bar chart, is a graph that visually represents categorical data using rectangular bars. Here are the key characteristics and uses of bar plots:

```python
import numpy as np
import matplotlib.pyplot as plt

# Provided data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color="skyblue")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company Profits")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color="salmon")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company Profits (Horizontal)")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
import numpy as np
import matplotlib.pyplot as plt

# Provided data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"], vert=False)
plt.xlabel("Values")
plt.ylabel("Boxes")
plt.title("Box Plot: Box 1 vs. Box 2")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```
