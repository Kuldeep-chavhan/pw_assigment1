Q2. What are the types of Probability distribution?


Probability distributions come in various forms, each describing how probabilities are distributed over the values of a random variable. They can be broadly classified into two categories: discrete and continuous distributions.

Discrete Distributions
Binomial Distribution: Models the number of successes in a fixed number of independent Bernoulli trials (e.g., flipping a coin multiple times).

Example: Number of heads in 10 coin flips.

Poisson Distribution: Represents the number of events occurring in a fixed interval of time or space, given a constant average rate and independence between events.

Example: Number of emails received per hour.

Geometric Distribution: Models the number of trials needed to get the first success in a sequence of independent Bernoulli trials.

Example: Number of coin flips until the first head appears.

Continuous Distributions
Normal (Gaussian) Distribution: The well-known bell curve that describes a symmetric distribution around the mean, characterized by the mean (Œº) and standard deviation (œÉ).

Example: Heights of people.

Exponential Distribution: Models the time between independent events occurring at a constant rate.

Example: Time between arrivals of buses.

Uniform Distribution: All outcomes are equally likely within a given range.

Example: Rolling a fair die (for a discrete uniform distribution), or any number in a range (for continuous).

Others
Bernoulli Distribution: Special case of binomial distribution with a single trial.

Example: Flipping a coin once.

Beta Distribution: Useful for modeling proportions and probabilities, defined on the interval [0, 1].

Example: Distribution of conversion rates.

Gamma Distribution: Generalization of the exponential distribution, often used to model waiting times.

Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import math

def normal_pdf(x, mean, std_dev):
    """Calculate the probability density function of a normal distribution."""
    exponent = math.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    pdf = (1 / (std_dev * math.sqrt(2 * math.pi))) * exponent
    return pdf

# Example usage:
x = 1.0
mean = 0.0
std_dev = 1.0
print(f"The PDF value is: {normal_pdf(x, mean, std_dev)}")
```

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 1000
p = 0.4

# Generate random sample
sample = np.random.binomial(1, p, n)

# Plot histogram
plt.hist(sample, bins=2, edgecolor='black', rwidth=0.8)
plt.title('Histogram of Binomial Distribution (n=1000, p=0.4)')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
from scipy.stats import poisson

def poisson_cdf(mean, point):
    """Calculate the CDF of a Poisson distribution at a given point."""
    cdf_value = poisson.cdf(point, mean)
    return cdf_value

# Example usage:
mean = 5
point = 3
print(f"The CDF value is: {poisson_cdf(mean, point)}")
```

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
import numpy as np

# Parameters
mean = 5
sample_size = 1000

# Generate random sample
sample = np.random.poisson(mean, sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution, the least frequent data appears in the tails of the distribution, farthest from the mean. The normal distribution is symmetrical about its mean, and it follows the bell curve shape, where the highest frequency of data points occurs around the mean.

Center (Mean): This is where the data points are most concentrated.

Tails: As you move away from the mean (in either direction towards the extremes), the frequency of data points decreases. These tails represent the least likely values in the distribution.
