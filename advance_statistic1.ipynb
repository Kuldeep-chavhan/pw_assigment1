Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

1. T-test vs Z-test: Key Differences
Sample Size:

T-test is typically used when the sample size is small (n < 30).
Z-test is used when the sample size is large (n ‚â• 30).
Population Standard Deviation:

T-test is used when the population standard deviation is unknown, and the sample standard deviation is used as an estimate.
Z-test is used when the population standard deviation is known.
Test Statistic:

The t-test uses the t-distribution, which is broader and accounts for the uncertainty with small samples.
The z-test uses the normal distribution (Z-distribution), which assumes the data is approximately normally distributed and works best with large sample sizes.
Distribution Shape:

T-distribution is wider, with heavier tails than the normal distribution, especially with smaller sample sizes.
Z-distribution is bell-shaped, similar to the normal distribution, and is used when the sample size is large enough for the Central Limit Theorem to apply.
2. Example Scenarios
T-test:

Scenario: A researcher wants to determine if the average height of a group of 20 students is different from the national average height of 170 cm. Since the sample size is small, and the population standard deviation is unknown, the researcher would use a t-test.
Here, you would use the one-sample t-test to test whether the mean height of the sample is significantly different from the known population mean of 170 cm.
Z-test:

Scenario: A company wants to test if their new manufacturing process results in a mean time to produce a product that is different from the industry standard of 50 minutes. They have data from a large sample of 1,000 production times, and they know the population standard deviation is 10 minutes.
In this case, since the sample size is large and the population standard deviation is known, the company would use a z-test (specifically, a one-sample z-test) to determine if the mean production time is significantly different from 50 minutes.

What is a Confidence Interval (CI)?
A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (e.g., mean or proportion) with a specified level of confidence. It reflects the uncertainty or variability inherent in sample-based estimates.

The confidence level (e.g., 95%) indicates the proportion of times the CI would contain the true parameter if repeated sampling were performed.
A higher confidence level results in a wider interval, while a lower confidence level results in a narrower interval.

Summary:

The relationship is inverse and proportional to the square root of the sample size.
