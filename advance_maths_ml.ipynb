Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.

```python
import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level=0.95):
    # Calculate the z-value for the desired confidence level
    z_value = {
        0.90: 1.645,
        0.95: 1.96,
        0.99: 2.576
    }
    z = z_value.get(confidence_level, 1.96)  # Default to 95% confidence if not specified

    # Calculate the margin of error
    margin_of_error = z * (sample_std_dev / math.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return (lower_bound, upper_bound)

# Example usage:
sample_mean = 50
sample_std_dev = 10
sample_size = 30
confidence_level = 0.95

ci = estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level)
print(f"Estimated Population Mean Confidence Interval: {ci}")
```

Hypothesis Testing
Hypothesis testing is a statistical method used to make decisions about the properties of a population based on sample data. It involves making an initial assumption (the null hypothesis) and testing whether there is enough evidence in the sample data to reject that assumption in favor of an alternative hypothesis.

Why is Hypothesis Testing Used?
Decision Making: It helps in making informed decisions about a population parameter based on sample data.

Validation of Assumptions: It allows researchers to test their theories and assumptions about the population.

Objectivity: It provides a structured framework for evaluating the evidence against the null hypothesis, reducing subjective bias.

Importance of Hypothesis Testing
Scientific Research: It's a cornerstone of scientific research, helping to validate experimental results and theories.

Quality Control: In industries, hypothesis testing is used for quality control and improvement processes.

Policy Making: It assists policymakers in making data-driven decisions.

Medical Trials: In healthcare, it is critical for determining the efficacy of new treatments and drugs.

Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.

```python
import numpy as np
from scipy import stats

def conduct_two_sample_ttest(sample1, sample2, alpha=0.05):
    # Calculate the means and standard deviations of the samples
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    std1 = np.std(sample1, ddof=1)
    std2 = np.std(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)

    # Calculate the test statistic and p-value
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=True)

    # Determine whether to reject the null hypothesis
    reject_null = p_value < alpha

    # Print the results
    print(f"Sample 1 Mean: {mean1}")
    print(f"Sample 2 Mean: {mean2}")
    print(f"T-Statistic: {t_statistic}")
    print(f"P-Value: {p_value}")
    print(f"Reject Null Hypothesis: {reject_null}")

    return t_statistic, p_value, reject_null

# Example usage:
sample1 = [70, 75, 80, 85, 90]
sample2 = [65, 70, 75, 80, 85]
alpha = 0.05

t_statistic, p_value, reject_null = conduct_two_sample_ttest(sample1, sample2, alpha)

if reject_null:
    print("There is significant evidence to suggest a difference in means between the two populations.")
else:
    print("There is no significant evidence to suggest a difference in means between the two populations.")
```

sample1 and sample2: Arrays containing the sample data from the two populations.

alpha: The significance level (default is 0.05).

ttest_ind: A function from scipy.stats to perform a two-sample t-test assuming equal variances.

t_statistic and p_value: The t-test statistic and the p-value from the test.

reject_null: A boolean indicating whether the null hypothesis is rejected.

This script calculates the test statistic and p-value for the two-sample t-test, and then determines whether there is enough evidence to reject the null hypothesis, which states that there is no difference between the population means.

Q8. Define p-value and explain its significance in hypothesis testing.

efinition of P-value
A p-value is the probability of obtaining a test statistic at least as extreme as the one observed, given that the null hypothesis is true. It quantifies the evidence against the null hypothesis. Essentially, it tells you how likely it is to observe the data assuming the null hypothesis is correct.

Significance of P-value in Hypothesis Testing
The p-value plays a crucial role in hypothesis testing. Here's how:

Evidence Against the Null Hypothesis:

A small p-value (typically ‚â§ 0.05) indicates strong evidence against the null hypothesis, so you reject the null hypothesis.

A large p-value (> 0.05) indicates weak evidence against the null hypothesis, so you fail to reject the null hypothesis.

Decision Making:

The p-value helps in making objective decisions about the null hypothesis. By comparing the p-value to a predetermined significance level (
ùõº
), you can decide whether to reject or fail to reject the null hypothesis.

Understanding Results:

The p-value provides a measure of the strength of the evidence. A lower p-value means that the observed data is less likely to occur under the null hypothesis, suggesting that the alternative hypothesis may be true.

Example
If you conduct a hypothesis test and get a p-value of 0.03 with a significance level of 0.05:

Since 0.03 < 0.05, you reject the null hypothesis.

This means there is sufficient evidence to support the alternative hypothesis.

Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.

```python
import numpy as np
from scipy import stats

def two_sample_ttest(sample1, sample2, alpha=0.05):
    # Calculate the means and standard deviations of the samples
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    std1 = np.std(sample1, ddof=1)
    std2 = np.std(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)

    # Perform the two-sample t-test
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=True)

    # Decision rule
    reject_null = p_value < alpha

    # Print results
    print(f"Sample 1 Mean: {mean1}")
    print(f"Sample 2 Mean: {mean2}")
    print(f"T-Statistic: {t_statistic}")
    print(f"P-Value: {p_value}")
    print(f"Reject Null Hypothesis: {reject_null}")

    return t_statistic, p_value, reject_null

# Example usage:
sample1 = [70, 75, 80, 85, 90]
sample2 = [65, 70, 75, 80, 85]
alpha = 0.05

t_statistic, p_value, reject_null = two_sample_ttest(sample1, sample2, alpha)

if reject_null:
    print("There is significant evidence to suggest a difference in means between the two populations.")
else:
    print("There is no significant evidence to suggest a difference in means between the two populations.")
```

Q11: What is Student‚Äôs t distribution? When to use the t-Distribution.

Student‚Äôs t-distribution, often simply called the t-distribution, is a probability distribution used in statistics that is symmetric and bell-shaped, much like the normal distribution, but with heavier tails. This means it has a higher probability of values far from the mean. The t-distribution becomes increasingly similar to the normal distribution as the degrees of freedom (df) increase.

The t-distribution is defined by the degrees of freedom, which are related to the sample size. The fewer the degrees of freedom, the fatter the tails, indicating a greater likelihood of extreme values.

When to Use the t-Distribution
The t-distribution is used in hypothesis testing and confidence interval estimation for small sample sizes or when the population standard deviation is unknown. Here are some common scenarios:

Small Sample Sizes: When the sample size is small (typically less than 30), the t-distribution is more appropriate than the normal distribution.

Unknown Population Standard Deviation: When the population standard deviation is not known, and instead, the sample standard deviation is used.

Hypothesis Testing: For example, in a t-test to compare the means of two independent samples (two-sample t-test) or to compare the mean of a single sample to a known value (one-sample t-test).

Confidence Intervals: When constructing confidence intervals for the mean of a population, especially with small sample sizes or unknown standard deviations.

Key Properties
Symmetric around the mean (zero).

Flatter and wider tails compared to the normal distribution.

Depends on the degrees of freedom (df).

As the sample size increases, the t-distribution approaches the normal distribution.

Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
pounds with a significance level of 0.01.

```python
import numpy as np
from scipy import stats

# Given data
sample_mean = 4.8
population_mean = 5
sample_std_dev = 0.5
sample_size = 25
alpha = 0.01

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))
print(f"T-Statistic: {t_statistic}")

# Degrees of freedom
df = sample_size - 1

# Calculate the p-value
p_value = stats.t.cdf(t_statistic, df)
print(f"P-Value: {p_value}")

# Decision rule
reject_null = p_value < alpha
print(f"Reject Null Hypothesis: {reject_null}")
```

Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.

```python
import numpy as np
from scipy import stats

# Given data
mean1 = 80
std_dev1 = 10
n1 = 30
mean2 = 75
std_dev2 = 8
n2 = 40
alpha = 0.01

# Calculate the pooled standard deviation
s_p = np.sqrt(((n1 - 1) * std_dev1**2 + (n2 - 1) * std_dev2**2) / (n1 + n2 - 2))

# Calculate the t-statistic
t_statistic = (mean1 - mean2) / (s_p * np.sqrt((1 / n1) + (1 / n2)))
print(f"T-Statistic: {t_statistic}")

# Degrees of freedom
df = n1 + n2 - 2

# Calculate the p-value
p_value = 2 * stats.t.sf(np.abs(t_statistic), df)
print(f"P-Value: {p_value}")

# Decision rule
reject_null = p_value < alpha
print(f"Reject Null Hypothesis: {reject_null}")
```
